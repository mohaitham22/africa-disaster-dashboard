```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Book1.csv')

# Check for complete rows (rows with no missing values)
complete_rows = df[df.notna().all(axis=1)]

print(f"Total rows in dataset: {len(df)}")
print(f"Complete rows (no missing values): {len(complete_rows)}")
print(f"\nPercentage of complete rows: {(len(complete_rows) / len(df)) * 100:.2f}%")

# Show which columns have the most missing values
print("\nMissing values per column:")
missing_counts = df.isnull().sum().sort_values(ascending=False)
print(missing_counts[missing_counts > 0].head(20))

# If there are complete rows, show a sample
if len(complete_rows) > 0:
    print("\nSample of complete rows:")
    print(complete_rows.head())
```
